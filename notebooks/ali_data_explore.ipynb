```python
def get_and_clean_data(f_path):
    df = pd.read_csv(f_path)
    df.index = df.id
    df.drop('id', axis=1, inplace=True)
    df.dropna(how='all', axis=[0, 1], inplace=True)
    return df
    
sec_2 = get_and_clean_data(file_paths['sec_2'])
sec_3 = get_and_clean_data(file_paths['sec_3'])
sec_4 = get_and_clean_data(file_paths['sec_4'])
sec_5 = get_and_clean_data(file_paths['sec_5'])
sec_7 = get_and_clean_data(file_paths['sec_7'])
sec_8 = get_and_clean_data(file_paths['sec_8'])
```

```python
# dealing with duplicated ids for s2 data

def nanmax(d_list):
    d_list_non_nan = d_list.dropna()
    if d_list_non_nan.shape[0] == 0:
        return np.NaN
    else:
        return max(list(d_list_non_nan))

agg_def = {
    'respondent': nanmax,
    'sex': nanmax,
    'plot_months': nanmax,
    'type_of_property': nanmax,
    'no_of_hhs': nanmax,
    'no_of_ppl': nanmax,
    'no_of_toilets': np.nansum,
    'add_toilets': nanmax,
    'water_source': nanmax,
    'water_source_other': nanmax,
    'type_of_toilet': nanmax,
    'type_of_property': nanmax,
    'latitude': nanmax,
    'longitude': nanmax,
    'accuracy': nanmax,
    'altitude': nanmax,
}

s2_agg = sec_2.groupby(sec_2.index).aggregate(agg_def)
```

```python
s2_agg = s2_agg.dropna(subset=['latitude', 'longitude'], axis=0, how='any')
s2_agg['no_of_toilets'] = s2_agg['no_of_toilets'].replace(0, np.NaN)

s2_agg['ppl_per_toilet'] = s2_agg['no_of_ppl']/s2_agg['no_of_toilets']

all_data_set = s2_agg.join(all_dat_4578, how='left')

print(s2_agg.shape)
print(all_data_set.shape)

all_data_set.head()
```

## Heatmaps

```python
def range_bound(x, curr_min, curr_max, new_min, new_max):
    return (((new_max - new_min)*(x-curr_min))/(curr_max - curr_min)) + new_min

def get_pltly_latlong_trace_hmap(df_in, size_category_name, cat_min, cat_max, size_min=1, size_max=30):
    df = df_in[['latitude', 'longitude', size_category_name]]
    df.dropna(how='any', axis=0, inplace=True)
    df = df[(df[size_category_name] >= cat_min) &
            (df[size_category_name] <= cat_max)]
    
    df['cat_scaled'] = df[size_category_name].apply(lambda x: range_bound(x, cat_min, cat_max, size_min, size_max))
    text_list = list(df[size_category_name])
    data = Scattermapbox(
            lat=df['latitude'],
            lon=df['longitude'],
            mode='markers',
            marker=Marker(
                size=df['cat_scaled'],
                color=df[size_category_name],
                colorscale='Jet',
                showscale=True,
                cmin=-5,
            ),
            text=text_list,
            name=size_category_name
            )
    return data



ppl_per_toilet_trace = get_pltly_latlong_trace_hmap(all_data_set, 'ppl_per_toilet', 0, 100, 1, 20)
when_next_full_mo_trace = get_pltly_latlong_trace_hmap(all_data_set, 'when_next_full_mo', 0, 180, 1, 15)


all_data = [ppl_per_toilet_trace, when_next_full_mo_trace]


access_token = 'pk.eyJ1IjoiYWxpc2lkZGlxIiwiYSI6ImNqZjQwdXExNzFib20zM2xyaHdzcWtjNmkifQ.kD2Isj8EgigfYVzVjXryFA'
layout = Layout(
    autosize=True,
    hovermode='closest',
    showlegend=True,
    mapbox=dict(
        style='streets',
#         style= 'mapbox://styles/mapbox/satellite-streets-v9',
        accesstoken=access_token,
        bearing=0,
        center=dict(
            lat=-15.434,
            lon=28.235
        ),
        pitch=0,
        zoom=13,
    ),
)

fig = dict(data=all_data, layout=layout)
offline.plot(fig, filename='../figures/{}_non_sat.html'.format('heatmaps'))
```
